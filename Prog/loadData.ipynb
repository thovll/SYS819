```python
import numpy as np
import os
import matplotlib.pyplot as plt

import nibabel as nib
```

```python
img2 = nib.load('mpr-1.nifti.img')
print(img2.shape)
data2 = img2.get_fdata()
print(data2.shape)
print(type(data2))

plt.subplot(1,3,1), plt.imshow(data2[:,101,:,0])
plt.subplot(1,3,2), plt.imshow(data2[101,:,:,0])
plt.subplot(1,3,3), plt.imshow(data2[:,:,101,0])
plt.colorbar()
```

```python
plt.hist(np.ravel(data2))
```
